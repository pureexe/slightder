```python
import torch 
import torchvision 

images = []
files = [(f,float(f)) for f in files]
files = sorted(files, key=lambda x: x[1])
files = [f[0] for f in files]
for idx, fname in enumerate(files):
    image = Image.open(f"../datasets/v1/scale-1_1/unsplash2000_ball/{fname}/image.png").resize((256,256))
    draw = ImageDraw.Draw(image)
    draw.text((10, 10),f"scale: {float(fname):.4f}",(255,255,255),font=font)
    images.append(torchvision.transforms.functional.pil_to_tensor(image))

images = torch.stack(images)
grid = torchvision.utils.make_grid(images, nrow=16)
grid = torchvision.transforms.ToPILImage()(grid)
#grid.save(f'output/unsplash2000_latent/chkpt10000_scaled{SCALE_IDX}.png')
#grid.save(f'output/unsplash2000_latent/chkpt9500_scaled{SCALE_IDX}.png')
grid.save(f'scale_line2.png')


    
```
