```python
def find_closest_index(array, value):
  """
  Finds the index of the element in the array that is closest to the given value.

  Args:
    array: A NumPy array.
    value: The value to compare to the elements in the array.

  Returns:
    The index of the element in the array that is closest to the given value.
  """
  # Calculate the absolute differences between the value and each element in the array.
  differences = np.abs(array - value)

  # Find the index of the element with the smallest absolute difference.
  return np.argmin(differences)
```

```python

for num_image in tqdm([2,4,8,12,23,46,91,180]):
    os.makedirs(f"output/shoe401_{num_image}_1.0_2.0/nearest_input/", exist_ok=True)
    files = os.listdir(f"datasets/scale1_2/shoe401_{num_image}_1.0_2.0")
    weights = np.array([float(file) for file in files])
    for scale_id, scale in enumerate(np.linspace(1, 2, 180)):
        index = find_closest_index(weights, scale)
        # read image from index and resize it to 256x256
        image = Image.open(f"datasets/scale1_2/shoe401_{num_image}_1.0_2.0/{files[index]}/shoe.png")
        image = image.resize((256, 256), Image.ANTIALIAS)
        image.save(f"output/shoe401_{num_image}_1.0_2.0/nearest_input/{scale_id:04d}.png")
        #os.system(f"cp datasets/scale1_2/shoe401_{num_image}_1.0_2.0/{files[index]} output/scale1_2/shoe401_{num_image}_1.0_2.0/{scale_id:04d}.png")
```
