```python
from matplotlib import pyplot as plt
import os
```

```python
files = os.listdir("../datasets/v1/scale-1_1/unsplash_cast_250")
```

```python
scales = [float(f) for f in files]
```

```python
import numpy as np 
scales = np.array(scales)
```

```python
# Define the number of bins (10 in this case)
num_bins = 10

# Calculate the histogram using np.histogram
counts, bins = np.histogram(scales, bins=num_bins)

# Plot the histogram
plt.bar(bins[:-1], counts, width=bins[1] - bins[0])  # Adjust bar width based on bin edges

# Add labels and title
plt.xlabel("left-right weight")
plt.ylabel("Count")
plt.title("Histogram of 250 image dataset (10 bins)")

# Display the plot
plt.show()
```
